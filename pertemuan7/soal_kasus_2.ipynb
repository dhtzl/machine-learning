```python
import pandas as pd
import matplotlib.pyplot as plt

p = ['gelas', 'piring', 'mangkok']
b = [100, 200, 300]

pb = {'product':p, 'barang':b}

data = pd.DataFrame(data=pb, columns=['product','barang'])
print(data)

plt = pd.DataFrame.plot(data)
data.groupby(['product']).sum().plot(kind='pie', y='barang', autopct='%1.0f%%')




```
