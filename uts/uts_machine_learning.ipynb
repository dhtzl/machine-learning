```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import sklearn
```

```python
p = ['kopiko', 'torabika', 'beng beng', 'choki-choki', 'teh pucuk', 'energen']
b = [52,39,78,13,78,29]

product_data = {'names':p, 'stok':b}

data = pd.DataFrame(data=product_data, columns=['names','stok'])
print(data)
```

```python

data.plot(x='names', y='stok', marker='o')
```

```python

plt = pd.DataFrame.plot(data)
data.groupby(['names']).sum().plot(kind='pie', y='stok', autopct='%1.1f%%')
```
