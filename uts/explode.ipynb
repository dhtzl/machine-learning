```python
import matplotlib.pyplot as plt

flavors = ('Teh pucuk','Beng beng','Choki-Choki', 'Kopiko','Torabika','energen')
votes   = (78, 78, 13, 52, 39, 29)
colors  = ('#8B4513', '#FFF8DC', '#93C572', '#E67F0D', '#D53032')
explode = (0, 0,  0.1, 0, 0, 0)

plt.pie(
    votes,
    labels=flavors,
    autopct='%1.1f%%',
    colors=colors,
    explode=explode,
)
plt.show()
```
