```python
import pandas as pd
import matplotlib.pyplot as plt

p = ['kopiko', 'torabika', 'beng beng', 'choki-choki', 'teh pucuk', 'energen']
b = [52,39,78,13,78,29]

pb = {'product':p, 'barang':b}

data = pd.DataFrame(data=pb, columns=['product','barang'])
print(data)

plt = pd.DataFrame.plot(data)
data.groupby(['product']).sum().plot(kind='pie', y='barang', autopct='%1.1f%%')




```
